```python
import math
import pandas as pd
import numpy as np
from sklearn import linear_model as lm
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
#Selecting the following predictors
#train_df[['MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'Street', 'LotShape', 'Utilities', 'BldgType', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'TotalBsmtSF', 'Heating', 'CentralAir', 'Electrical', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'KitchenQual', 'TotRmsAbvGrd', 'Functional', 'Fireplaces', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF','EnclosedPorch', 'PoolArea']].isnull()
```

```python
# Replace string values with numerical scale

# MSZoning
train_df['MSZoning'].replace('A',0, inplace=True)
train_df['MSZoning'].replace('C',1, inplace=True)
train_df['MSZoning'].replace('C (all)',1, inplace=True)
train_df['MSZoning'].replace('FV',2, inplace=True)
train_df['MSZoning'].replace('I',3, inplace=True)
train_df['MSZoning'].replace('RH',4, inplace=True)
train_df['MSZoning'].replace('RL',5, inplace=True)
train_df['MSZoning'].replace('RP',6, inplace=True)
train_df['MSZoning'].replace('RM',7, inplace=True)

# Street
train_df['Street'].replace('Grvl',0, inplace=True)
train_df['Street'].replace('Pave',1, inplace=True)

# LotShape
train_df['LotShape'].replace('Reg',0, inplace=True)
train_df['LotShape'].replace('IR1',1, inplace=True)
train_df['LotShape'].replace('IR2',2, inplace=True)
train_df['LotShape'].replace('IR3',3, inplace=True)

# Utilities
train_df['Utilities'].replace('AllPub',0, inplace=True)
train_df['Utilities'].replace('NoSewr',1, inplace=True)
train_df['Utilities'].replace('NoSeWa',2, inplace=True)
train_df['Utilities'].replace('ELO',3, inplace=True)

# BldgType
train_df['BldgType'].replace('1Fam',0, inplace=True)
train_df['BldgType'].replace('2FmCon',1, inplace=True)
train_df['BldgType'].replace('2fmCon',1, inplace=True)
train_df['BldgType'].replace('Duplx',2, inplace=True)
train_df['BldgType'].replace('Duplex',2, inplace=True)
train_df['BldgType'].replace('Twnhs',3, inplace=True)
train_df['BldgType'].replace('TwnhsE',3, inplace=True)
train_df['BldgType'].replace('TwnhsI',4, inplace=True)

# Heating
train_df['Heating'].replace('Floor',0, inplace=True)
train_df['Heating'].replace('GasA',1, inplace=True)
train_df['Heating'].replace('GasW',2, inplace=True)
train_df['Heating'].replace('Grav',3, inplace=True)
train_df['Heating'].replace('OthW',4, inplace=True)
train_df['Heating'].replace('Wall',5, inplace=True)

# CentralAir
train_df['CentralAir'].replace('N',0, inplace=True)
train_df['CentralAir'].replace('Y',1, inplace=True)

# Electrical
train_df['Electrical'].replace('SBrkr',0, inplace=True)
train_df['Electrical'].replace('FuseA',1, inplace=True)
train_df['Electrical'].replace('FuseF',2, inplace=True)
train_df['Electrical'].replace('FuseP',3, inplace=True)
train_df['Electrical'].replace('Mix',4, inplace=True)

# KitchenQual
train_df['KitchenQual'].replace('Ex',0, inplace=True)
train_df['KitchenQual'].replace('Gd',1, inplace=True)
train_df['KitchenQual'].replace('TA',2, inplace=True)
train_df['KitchenQual'].replace('Fa',3, inplace=True)
train_df['KitchenQual'].replace('Po',4, inplace=True)

# Functional
train_df['Functional'].replace('Typ',0, inplace=True)
train_df['Functional'].replace('Min1',1, inplace=True)
train_df['Functional'].replace('Min2',2, inplace=True)
train_df['Functional'].replace('Mod',3, inplace=True)
train_df['Functional'].replace('Maj1',4, inplace=True)
train_df['Functional'].replace('Maj2',5, inplace=True)
train_df['Functional'].replace('Sev',6, inplace=True)
train_df['Functional'].replace('Sal',7, inplace=True)
```

```python
# Impute missing values
avgLotFrontage = train_df['LotFrontage'].mean()
train_df['LotFrontage'].fillna(avgLotFrontage, inplace=True)

avgElectrical = train_df['Electrical'].mean()
train_df['Electrical'].fillna(avgElectrical, inplace=True)
```

```python
# Check unique values
train_df['MSZoning'].unique()
```

```python
# Sets
X = train_df[['MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'Street', 'LotShape', 'Utilities', 'BldgType', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'TotalBsmtSF', 'Heating', 'CentralAir', 'Electrical', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'KitchenQual', 'TotRmsAbvGrd', 'Functional', 'Fireplaces', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF','EnclosedPorch', 'PoolArea']]

#Identify Y - Segregate Y
Y = train_df['SalePrice']

#Create a linear regression object - Not correct for this dataset. But I try to fit one for learning purposes
lmod = lm.LinearRegression()
lmod.fit(X, Y)
```

```python
#Print parameters
print('Intercept : ',lmod.intercept_)
pd.DataFrame(list(zip(X.columns, lmod.coef_)), columns=['features','estimatedCoeffs'])
```

```python
# Get the fitted values
y_fittedVals = lmod.predict(X)

# Get the MSE
print("Mean squared error: %.2f"
      % mean_squared_error(Y, y_fittedVals))

print('R-Square: %.2f' % r2_score(Y, y_fittedVals))

#y_fittedVals
```

## Doing the same using Stats model
[Link](http://www.statsmodels.org/devel/generated/statsmodels.regression.linear_model.OLS.html)

```python
import statsmodels.api as sm
```

```python
# add_constant has to be done by the user. Otherwise it will not add a intercept to the model
X2 = sm.add_constant(X)
est = sm.OLS(Y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
# From the above summary we see that the following predictors are not significant at p=0.05 level
# Drop these columns from the list and then redo the regression again
X3 = X.drop(columns=['MSSubClass','MSZoning','Street','LotShape','Utilities','BldgType','YearRemodAdd','Heating','CentralAir','Electrical','1stFlrSF','2ndFlrSF','GrLivArea','TotRmsAbvGrd','OpenPorchSF','EnclosedPorch','PoolArea'])

# add_constant has to be done by the user. Otherwise it will not add a intercept to the model
X2 = sm.add_constant(X3)
est = sm.OLS(Y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
#But R2 drops well and hence I checked the individual predictors again and '1stFlrSF','2ndFlrSF' have significant influence on the model. 
# So, when all the parameters are taken together then they seem to be less influential because they are highly correlated with GrLivArea
# Hence we reintroduce them into the model and remove GrLivArea from the model and drop LotFrontage

# From the above summary we see that the following predictors are not significant at p=0.05 level
# Drop these columns from the list and then redo the regression again
X3 = X.drop(columns=['MSSubClass','MSZoning','Street','LotShape','Utilities','BldgType','YearRemodAdd','Heating','CentralAir','Electrical','GrLivArea','TotRmsAbvGrd','OpenPorchSF','EnclosedPorch','PoolArea', 'LotFrontage'])

# add_constant has to be done by the user. Otherwise it will not add a intercept to the model
X2 = sm.add_constant(X3)
est = sm.OLS(Y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
# So we finally get the model with R2 of 0.796 which is pretty strong
# We can check for linear fit, non-constant variance etc. But we leave it for now and proceed to the test dataset and 
# check for accuracy

#Check to see if there are any null values in the test dataset
#test_df[['LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'KitchenQual', 'Functional', 'Fireplaces', 'GarageArea', 'WoodDeckSF']].isnull().values.any()

for ind in ['LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'KitchenQual', 'Functional', 'Fireplaces', 'GarageArea', 'WoodDeckSF']:
    print("Val: " + str(ind) + ", " + str(test_df[ind].isnull().values.any()))
```

```python
# Yes there are Null values but first we replace string categories with numrical values
# Replace string values with numerical scale
# KitchenQual
test_df['KitchenQual'].replace('Ex',0, inplace=True)
test_df['KitchenQual'].replace('Gd',1, inplace=True)
test_df['KitchenQual'].replace('TA',2, inplace=True)
test_df['KitchenQual'].replace('Fa',3, inplace=True)
test_df['KitchenQual'].replace('Po',4, inplace=True)

# Functional
test_df['Functional'].replace('Typ',0, inplace=True)
test_df['Functional'].replace('Min1',1, inplace=True)
test_df['Functional'].replace('Min2',2, inplace=True)
test_df['Functional'].replace('Mod',3, inplace=True)
test_df['Functional'].replace('Maj1',4, inplace=True)
test_df['Functional'].replace('Maj2',5, inplace=True)
test_df['Functional'].replace('Sev',6, inplace=True)
test_df['Functional'].replace('Sal',7, inplace=True)

# Impute values now
avgTotBsmtSF = test_df['TotalBsmtSF'].mean()
test_df['TotalBsmtSF'].fillna(avgTotBsmtSF, inplace=True)

avgGargArea = test_df['GarageArea'].mean()
test_df['GarageArea'].fillna(avgGargArea, inplace=True)

avgKiQual = math.floor(test_df['KitchenQual'].mean())
test_df['KitchenQual'].fillna(avgKiQual, inplace=True)

avgFunc = math.floor(test_df['Functional'].mean())
test_df['Functional'].fillna(avgFunc, inplace=True)
```

```python
X_test = sm.add_constant(test_df[['Id', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'KitchenQual', 'Functional', 'Fireplaces', 'GarageArea', 'WoodDeckSF']])

X_test.columns

# Predict the test results
Y_prob_pred = est2.predict(X_test[['const','LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'KitchenQual', 'Functional', 'Fireplaces', 'GarageArea', 'WoodDeckSF']])
```

```python
X_test['SalePrice'] = Y_prob_pred
X_test.head()
```
